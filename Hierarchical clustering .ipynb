```python
iris = datasets.load_iris()
data = iris.data
```

```python
x_features = iris.data
y_labels = iris.target
```

```python
model = AgglomerativeClustering(linkage="ward",n_clusters=3)
```

```python
model.fit(x_features)
predicted_labels = model.labels_
```

```python
predicted_labels
```

```python
linkage_type = 'ward'
normalized_x = (data - np.mean(data, axis=0)) / np.std(data, axis=0)
```

```python
linkage_matrix = linkage(normalized_x, linkage_type)
```

```python
distance_matrix = linkage_matrix[:, 2].reshape(-1, 1)
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(22,18))
dendrogram(linkage_matrix)
plt.show()
```

```python
plt.figure(figsize=(22, 18))
dendrogram(
    linkage_matrix,
    leaf_rotation=90,  # Rotates the x-axis labels
    leaf_font_size=12,  # Font size for the x-axis labels
    above_threshold_color='cyan',  # Color for links above the threshold
    color_threshold=0.7 * max(linkage_matrix[:, 2]),  # Color threshold
    truncate_mode='lastp',  # Truncates the dendrogram
    p=30  # Number of last merges to show
)
plt.title('Hierarchical Clustering Dendrogram (iris Dataset)', fontsize=20)
plt.xlabel('Sample index or (Cluster size)', fontsize=16)
plt.ylabel('Distance', fontsize=16)
plt.grid(True)
plt.show()
```
